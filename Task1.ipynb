```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url="http://bit.ly/w-data"
s_data=pd.read_csv(url)
print("Data imported successfully")
s_data.head(10)
```

```python
print(s_data.shape)
```

```python
#plotting the distribution of the scores
s_data.plot(x='Hours', y='Scores', style='o')
plt.title("Hours vs Score")
plt.xlabel("Hours of Study")
plt.ylabel("Score Achieved")
plt.show()
```

```python
#attribute
x=s_data.iloc[:, :-1].values
y=s_data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)

print("Training complete.")
```

```python
import matplotlib.pyplot as plt

# Plotting the regression line
line = regressor.coef_*x+regressor.intercept_

# Plotting for the test data
plt.scatter(x, y)
plt.plot(x, line)
plt.show()
```

```python
#making predictions
print(x_test)   # testing data - in hours
y_pred = regressor.predict(x_test)  # predicting the scores
```

```python
#Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x_train =  s_data.iloc[:, :-1].values
y_train =  s_data.iloc[:, 1].values

# Create and train the regressor (assuming you have your own data)
regressor = LinearRegression()
regressor.fit(x_train, y_train)
# Fit the model to your data

# Define the number of hours you want to predict for
hours = np.array([[9.25]])  # Convert to a 2D array

# Make a prediction
own_pred = regressor.predict(hours)

# Print the results
print("No of Hours = {}".format(hours[0][0]))
print("Predicted Score = {}".format(own_pred[0]))
```
